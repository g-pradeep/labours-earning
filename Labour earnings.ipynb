```python
df_num=df[['Age', 'Nodeg','Earnings_1974', 'Earnings_1975', 'Earnings_1978']]
```

```python
df_num.head()
```

```python
df_dummy=pd.get_dummies(df[['Race', 'Hisp', 'MaritalStatus', 'Eduacation']])
```

```python
df_dummy.head()
```

```python
df_combo=pd.concat([df_num,df_dummy],axis=1)
```

```python
df_combo.head()
```

```python
X = df_combo[['Age', 'Earnings_1974', 'Earnings_1975', 'Race_NotBlack', 'Race_black', 
                         'Hisp_NotHispanic', 'Hisp_hispanic','MaritalStatus_Married', 
                         'MaritalStatus_NotMarried', 'Eduacation_HighSchool', 'Eduacation_Intermediate',
                         'Eduacation_LessThanHighSchool', 'Eduacation_PostGraduate', 'Eduacation_graduate']]
y = df['Earnings_1978']
```

```python
pip install sklearn
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,y, train_size=0.5 , random_state=100)
```

```python
from sklearn.linear_model import LinearRegression
LM=LinearRegression()
```

```python
LM.fit(X_train,Y_train)
```

```python
corrMatrix= df.corr()
import seaborn as sn
sn.heatmap(corrMatrix, annot=True)
```

```python
print(LM.intercept_)
```

```python
coeff_df = pd.DataFrame(LM.coef_,X_test.columns,columns=['Coefficient'])
coeff_df
```

#Predictions

```python
# Making predictions using the model
y_pred = LM.predict(X_test)
```

```python
#Model Performance Metrics
```

```python
from IPython.display import Image
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(Y_test, y_pred)
r_squared = r2_score(Y_test, y_pred)
```

```python
from math import sqrt
rmse = sqrt(mse)
```

```python
print('Mean_Squared_Error :' ,mse)
print('Root_Mean_Squared_Error :' ,rmse)
print('r_square_value :',r_squared)
```

Checking for P-value Using STATSMODELS

```python
import statsmodels.api as sm
X_train_sm = X_train
#Unlike SKLearn, statsmodels don't automatically fit a constant, 
#so you need to use the method sm.add_constant(X) in order to add a constant. 
X_train_sm = sm.add_constant(X_train_sm)
# create a fitted model in one line
LM_1 = sm.OLS(Y_train,X_train_sm).fit()

# print the coefficients
LM_1.params
```

```python
print(LM_1.summary())
```
